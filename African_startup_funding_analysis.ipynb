```python
#Importing required libraries

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%reload_ext autoreload
%autoreload 2

import seaborn as sns

import matplotlib.ticker as mticker
```

```python
company_df = pd.read_excel(r'C:\Users\Wande\Desktop\Portfolio_Projects\Africa_tech_startup_funding_analysis\african_startups.xlsx')
company_df.head()
```

```python
#.1 Most startups that were funded are based in Nigeria

plt.figure(figsize=(12, 8))
sns.countplot(company_df.Location)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
#2. Male founders received more investment than female founded company
```

```python
#Removing space from the values in 'Female_co_founder_CEO' column 

company_df['Female_co_founder_CEO'] = company_df['Female_co_founder_CEO'].apply(lambda x: x.replace(' ', ''))
company_df.head()
```

```python

sns.countplot(company_df.Female_co_founder_CEO)
```

```python
#3. startups founded by international founders are more likely to be funded

int_founder = pd.DataFrame(company_df['Intnl founder?'].value_counts())
int_founder.head()
```

```python
labels = ['No', 'Yes']
colors = ['red', 'green'] 
plt.figure(figsize=(6,6))
_,_, autotexts = plt.pie(int_founder, autopct='%.0f%%', labels = labels, colors = colors )
for autotext in autotexts:
    autotext.set_color('white')
plt.show()
```

```python
#4. Financial technology company received the highest amount of fund in total

#Checking rows for NaN
null_df = company_df[company_df.isna().any(axis=1)]
null_df.head()
```

```python
#dropping all rows containing NaN

company_df = company_df.dropna()
company_df.head()
```

```python
#Confirming no NaN value
null_df = company_df[company_df.isna().any(axis=1)]
null_df.head()
```

```python
top_funded = company_df.groupby('Sector').sum()['Amount (US$)'].reset_index()
top_funded
```

```python
top_ten = top_funded.sort_values(['Amount (US$)'], ascending = False).head(10)
top_ten
```

```python
plt.figure(figsize = (15, 8))
plt.bar(top_ten['Sector'], top_ten['Amount (US$)'], color='purple')

plt.ticklabel_format(style='plain', useOffset=False, axis='y') #To remove scientific notation e.g. le6 

plt.xlabel('Sector')
plt.ylabel('Amount (US$)')
plt.show()
```
